# Fingerprint Exploration Notebook

This notebook is used for exploratory data analysis and visualization of fingerprint data. It includes code for loading datasets, visualizing fingerprints, and testing preprocessing methods.

```python
# Import necessary libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
from src.preprocess import enhance_image, binarize_image
from src.fingerprint import Fingerprint

# Load fingerprint image
image_path = 'path/to/fingerprint/image.jpg'
fingerprint_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Display original fingerprint image
plt.imshow(fingerprint_image, cmap='gray')
plt.title('Original Fingerprint')
plt.axis('off')
plt.show()

# Enhance the image
enhanced_image = enhance_image(fingerprint_image)

# Display enhanced fingerprint image
plt.imshow(enhanced_image, cmap='gray')
plt.title('Enhanced Fingerprint')
plt.axis('off')
plt.show()

# Binarize the image
binarized_image = binarize_image(enhanced_image)

# Display binarized fingerprint image
plt.imshow(binarized_image, cmap='gray')
plt.title('Binarized Fingerprint')
plt.axis('off')
plt.show()

# Initialize Fingerprint class
fingerprint = Fingerprint()

# Extract features from the fingerprint
features = fingerprint.extract_features(binarized_image)
print('Extracted Features:', features)
```
